```python
Bootstrap (Bagging) refers to random sampling with replacement. Bootstrap allows us to better understand the bias and the variance with the dataset. Bootstrap involves random sampling of small subset of data from the dataset.

It is a general procedure that can be used to reduce the variance for those algorithm that have high variance, typically decision trees. Bagging makes each model run independently and then aggregates the outputs at the end without preference to any model.



Boosting (originally called hypothesis boosting) refers to any Ensemble method that
can combine several weak learners into a strong learner. The general idea of most
boosting methods is to train predictors sequentially, each trying to correct its predecessor.

Random forest is a bagging technique and not a boosting technique
```

```python
# Perform Grid-Search
def rfr_model(X, y):
    gsc = GridSearchCV(
        estimator=RandomForestRegressor(),
        param_grid={
            'max_depth': range(3,7),
            'n_estimators': (10, 50, 100, 1000),
        },
        cv=5, scoring='neg_mean_squared_error', verbose=0, n_jobs=-1)
    
    grid_result = gsc.fit(X, y)
    best_params = grid_result.best_params_
    
    
    return best_params
```

```python
rfr_model(X_train,y)
```

```python
rfr = RandomForestRegressor(max_depth=6, n_estimators=1000,random_state=False)
rfr.fit(X_train,y)
y_pred = rfr.predict(X_test)
#scores = cross_val_score(rfr, X, y, cv=10, scoring='neg_mean_squared_error')
res = pd.DataFrame({'Id': range(1461,2920), 'SalePrice': y_pred})
res.tail()
res.to_csv('results_rfr.csv')
```

XGBoost : Better Performance, Less computing, Handles missing data, Regularization, Cross Validation Built in

```python
import xgboost as xgb
# Perform Grid-Search
def xgbr_model(X, y):
    gsc = GridSearchCV(
        estimator=xgb.XGBRegressor(),
        param_grid={
            'max_depth': range(3,7),
            'n_estimators': (10, 50, 100, 500),
            'alpha': range(0,5),
            'colsample_bytree': (0.1,0.3,0.5,0.7,1),
        })
    
    grid_result = gsc.fit(X, y)
    best_params = grid_result.best_params_
    
    
    return best_params
```

```python
xgbr_model(X_train_all,y)
```

```python
xgb_reg= xgb.XGBRegressor(alpha=0,colsample_bytree=0.1,max_depth=4,n_estimators=500)
xgb_reg.fit(X_train_all,y)
y_pred=xgb_reg.predict(X_test_all)
```
